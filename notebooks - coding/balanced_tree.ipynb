Given a sorted array. Write a function that creates a Balanced Binary Search Tree using array elements.

```
class Node:
    def __init__(self,val):
        self.val = val
        self.left = None
        self.right = None


class BST:
    def __init__(self):
        self.root = None

    def in_order(self, root_):
        if root_ is None:
            return
        self.in_order(root_.left)
        print(root_.val)
        self.in_order(root_.right)

    def pre_order(self, root_):
        if root_ is None:
            return
        print(root_.val)
        self.pre_order(root_.left)
        self.pre_order(root_.right)

    def make_balanced(self, input_):
        mid = len(input_) // 2
        if self.root is None:
            self.root = Node(input_[mid])
        self._make_balanced(input_, self.root)

    def _make_balanced(self, input_, root_):
        mid = len(input_) // 2
        if len(input_) <= 1:
            return

        input_left = input_[0:mid]
        input_right = input_[mid+1:]
        if len(input_left) != 0:
            if root_.left is None:
                mid_left = len(input_left) // 2
                root_.left = Node(input_left[mid_left])
            self._make_balanced(input_left, root_.left)
        if len(input_right) != 0:
            if root_.right is None:
                mid_right = len(input_right) // 2
                root_.right = Node(input_right[mid_right])
            self._make_balanced(input_right, root_.right)
```
