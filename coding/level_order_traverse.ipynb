```
import queue

class Node:
  def __init__(self, val):
    self.val = val
    self.right = None
    self.left = None

def level_order_traverse(root):
  if not root:
    return []
  traverse_nodes = queue.Queue()
  traverse_nodes.put((root, 1))

  result = []
  while traverse_nodes.empty() is False:
    front_elem = traverse_nodes.get(0)
    if len(result) < front_elem[1]:
      result += [[front_elem[0].val]]
    else:
      result[front_elem[1] - 1].append(front_elem[0].val)

    if front_elem[0].left:
      traverse_nodes.put((front_elem[0].left, front_elem[1]+1))

    if front_elem[0].right:
      traverse_nodes.put((front_elem[0].right, front_elem[1]+1))
      
  return result
  

if __name__ == "__main__":

  root = Node(5)
  root.left = Node(6)
  root.right = Node(7)
  root.left.right = Node(10)
  root.left.left = Node(17)
  root.right.right = Node(4)


  result = level_order_traverse(root)
  print(result)
```
