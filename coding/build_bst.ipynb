```
class Node:
  def __init__(self, val):
    self.val = val
    self.left = None
    self.right = None

class BST:
  def __init__(self):
    self.root = None

  def insert(self, val):
    if self.root is None:
      self.root = Node(val)
    else:
      self.insert_node(self.root, val)
  
  def insert_node(self, node, val):
    if node is None:
      node = Node(val)
      return
    if val > node.val:
      if node.right is None:
        node.right = Node(val)
      else:
        self.insert_node(node.right, val)
    else:
      if node.left is None:
        node.left = Node(val)
      else:
        self.insert_node(node.left, val)

  def find(self, val):
    return self.find_node(self.root, val)

  def find_node(self, node, val):
    if node == None:
      return False
    
    if node.val == val:
      return True

    if val > node.val:
      node = node.right
    else:
      node = node.left

    return self.find_node(node, val)

  def in_order(self):
    self.traverse_in_order(self.root)

  def traverse_in_order(self, node):
    if node is None:
        return
    self.traverse_in_order(node.left)
    print(node.val)
    self.traverse_in_order(node.right)

  def height(self):
    return self.height_count(self.root)

  def height_count(self, node):
    if node is None:
      return 0 
    else:
      return 1 + max(self.height_count(node.right), self.height_count(node.left))
```
