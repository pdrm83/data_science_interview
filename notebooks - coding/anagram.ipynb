```
def anagram(s1, s2):

  s1 = s1.replace(' ','').lower()
  s2 = s2.replace(' ','').lower()

  if len(s1) != len(s2):
    return False
  
  count = {}

  for letter in s1:
    if letter in count:
      count[letter] += 1
    else:
      count[letter] = 1

  for letter in s2:
    if letter in count:
      count[letter] -= 1
    else:
      count[letter] = 1

  for item in count:
    if count[item] != 0:
      return False

  return True
```

```
if __name__ == "__main__":
  s1 = 'god'
  s2 = ' a g d'
  print(anagram('clint eastwood', 'old west action'))
  print(anagram('pedram', 'per MAD'))
  print(anagram('salam', 'sala'))
```
