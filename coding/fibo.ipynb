```
def fibo(num, states):
  if num in [1, 2]:
    states[num - 1] = 1
    return 1

  if states[num - 1] != 0:
    return states[num - 1]

  states[num - 1] = fibo(num - 1, states) + fibo(num - 2, states)
  return states[num - 1]


if __name__ == "__main__":
  num = 6
  states = [0 for i in range(num)]
  print(fibo(num, states))
  
```
