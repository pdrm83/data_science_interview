```
def find_substring(S, T):

  if len(T) > len(S):
    print('It is not possible!')

  pointer_2 = 0
  results = {}

  for pointer_1 in range(len(S)):
    states = [0 for i in range(len(T))]
    while 0 in states:
      print(states)
      print(pointer_2)
      if S[pointer_2] in T:
        states[T.index(S[pointer_2])] += 1
      pointer_2 += 1

    results[str(pointer_1)] = S[pointer_1:pointer_2]
    print(results)
    
    
  return results
```
